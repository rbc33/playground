We are playing the Guessing Game. The game will work as follows:

1. I pick a number between `1` and `n`.
2. You guess a number.
3. If you guess the right number, **you win the game**.
4. If you guess the wrong number, then I will tell you whether the number I picked is **higher or lower**, and you will continue guessing.
5. Every time you guess a wrong number `x`, you will pay `x` dollars. If you run out of money, **you lose the game**.

Given a particular `n`, return _the minimum amount of money you need to **guarantee a win regardless of what number I pick**_.

**Example 1:**

![](https://assets.leetcode.com/uploads/2020/09/10/graph.png)

**Input:** n = 10
**Output:** 16
**Explanation:** The winning strategy is as follows:
- The range is \[1,10\]. Guess 7.
    - If this is my number, your total is $0. Otherwise, you pay $7.
    - If my number is higher, the range is \[8,10\]. Guess 9.
        - If this is my number, your total is $7. Otherwise, you pay $9.
        - If my number is higher, it must be 10. Guess 10. Your total is $7 + $9 = $16.
        - If my number is lower, it must be 8. Guess 8. Your total is $7 + $9 = $16.
    - If my number is lower, the range is \[1,6\]. Guess 3.
        - If this is my number, your total is $7. Otherwise, you pay $3.
        - If my number is higher, the range is \[4,6\]. Guess 5.
            - If this is my number, your total is $7 + $3 = $10. Otherwise, you pay $5.
            - If my number is higher, it must be 6. Guess 6. Your total is $7 + $3 + $5 = $15.
            - If my number is lower, it must be 4. Guess 4. Your total is $7 + $3 + $5 = $15.
        - If my number is lower, the range is \[1,2\]. Guess 1.
            - If this is my number, your total is $7 + $3 = $10. Otherwise, you pay $1.
            - If my number is higher, it must be 2. Guess 2. Your total is $7 + $3 + $1 = $11.
The worst case in all these scenarios is that you pay $16. Hence, you only need $16 to guarantee a win.

**Example 3:**

**Input:** n = 2
**Output:** 1
**Explanation:** There are two possible numbers, 1 and 2.
- Guess 1.
    - If this is my number, your total is $0. Otherwise, you pay $1.
    - If my number is higher, it must be 2. Guess 2. Your total is $1.
The worst case is that you pay $1.

**Constraints:**

- `1 <= n <= 200`

```python
def getMoneyAmount(self, n: int) -> int:
    # dp[i][j] representa el mínimo dinero necesario para ganar
    # cuando el número está entre i y j
    dp = [[0] * (n + 1) for _ in range(n + 1)]
    
    # Para rangos de longitud 2 o más
    # length es el tamaño del rango que estamos considerando
    for length in range(2, n + 1):
        # start es el inicio del rango
        # n - length + 2 asegura que no nos pasemos del final
        for start in range(1, n - length + 2):
            # end es el final del rango
            end = start + length - 1
            min_cost = float('inf')
            
            # probamos cada número en el rango como adivinanza
            for guess in range(start, end + 1):
                cost = guess + max(
                    # coste del subrango izquierdo (si existe)
                    dp[start][guess-1] if guess > start else 0,
                    # coste del subrango derecho (si existe)
                    dp[guess+1][end] if guess < end else 0
                )
                min_cost = min(min_cost, cost)
            
            dp[start][end] = min_cost
```

Para n = 10:

Primero calculamos para rangos pequeños
Luego para rangos más grandes
En cada rango, probamos cada número como primera adivinanza
Por ejemplo, para [1,3]:

Si adivinamos 1: costo = 1 + subrango[2,3]
Si adivinamos 2: costo = 2 + max(subrango[1,1], subrango[3,3])
Si adivinamos 3: costo = 3 + subrango[1,2]
La clave es:

Usar DP para evitar recálculos
Para cada rango, probar todas las posibles primeras adivinanzas
Tomar el mínimo coste que garantiza ganar en el peor caso


Para n=3, paso a paso:

Primero length=2 (rangos de 2 números):

[1,2]:
Si adivino 1: coste = 1 + [2,2] = 1 + 0 = 1
Si adivino 2: coste = 2 + [1,1] = 2 + 0 = 2
dp[1][2] = min(1,2) = 1
[2,3]:
Si adivino 2: coste = 2 + [3,3] = 2 + 0 = 2
Si adivino 3: coste = 3 + [2,2] = 3 + 0 = 3
dp[2][3] = min(2,3) = 2
Luego length=3 [1,3]:

Si adivino 1: coste = 1 + dp[2][3] = 1 + 2 = 3
Si adivino 2: coste = 2 + max(dp[1][1], dp[3][3]) = 2 + 0 = 2
Si adivino 3: coste = 3 + dp[1][2] = 3 + 1 = 4
dp[1][3] = min(3,2,4) = 2
La respuesta es 2, que significa:

Adivina 2 primero (pierdes 2)
Si es muy alto, adivina 1 (gratis porque solo queda una opción)
Si es muy bajo, adivina 3 (gratis porque solo queda una opción)
Este es el mínimo dinero que garantiza ganar sin importar qué número sea.

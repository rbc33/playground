
#  Move Zeroes

Given an integer array nums, move all 0's to the end of it while maintaining 
the relative order of the non-zero elements.

#### Input: nums = [0,1,0,3,12]
#### Output: [1,3,12,0,0]
#### Move all zeros to end maintaining relative order of non-zero elements

```python
def movZeros(arr):
    i = 0
    while i < len(arr):
        if arr[i] == 0:
            arr.pop(i)
            arr.append(0)
            if arr[i] == 0:  # Necesario para evaluar el nuevo elemento
                continue
        i += 1
    return arr
```
